```python
# Add each prefix and suffix to remove to a list.
prefixes_suffixes = ["Dr. ", "Mr. ","Ms. ", "Mrs. ", "Miss ", " MD", " DDS", " DVM", " PhD"]
```

```python
# Iterate through the words in the "prefixes_suffixes" list and replace them with an empty space, "".
for word in prefixes_suffixes:
    students_data_df["student_name"] = students_data_df["student_name"].str.replace(word,"")
```

```python
# combine the data in a single dataset
school_data_complete_df = pd.merge(students_data_df, school_data_df, on = ['school_name','school_name'])
school_data_complete_df.head()
```

```python
# get the total number of students
student_count = school_data_complete_df['student_name'].count()
student_count
```

```python
# get the number of schools unique() method
school_count = len(school_data_complete_df['school_name'].unique())
school_count
```

```python
# calculate the total budget
total_budget = school_data_df['budget'].sum()
total_budget
```

```python
# get the avg math score
avg_math_score = school_data_complete_df['math_score'].mean()
avg_math_score
```

```python
# get the avg reading score
avg_reading_score = school_data_complete_df['reading_score'].mean()
avg_reading_score
```

```python
passing_math = school_data_complete_df['math_score'] >= 70
passing_reading = school_data_complete_df['reading_score'] >= 70
```

```python
# get all the students who are passig math in a new DF
passing_math = school_data_complete_df[school_data_complete_df['math_score'] >= 70]
passing_math.head()
```

```python
# get all the students who are passig reading in a new DF
passing_reading = school_data_complete_df[school_data_complete_df['reading_score'] >= 70]
passing_reading.head()
```

```python
# calculate the number of students passing math
passing_math_count = passing_math['student_name'].count()

#calculate the number of students passing reading
passing_reading_count = passing_reading['student_name'].count()

print(passing_math_count)
print(passing_reading_count)
```

```python
# calculate percentage of students who passed math
perct_passing_math = passing_math_count/float(student_count) * 100

# calculate perct of students passing reading
perct_passing_reading = passing_reading_count/float(student_count) * 100

print(perct_passing_math)
print(perct_passing_reading)
```

```python
# calc # of students who passed both math and reading

passing_math_reading = school_data_complete_df[(school_data_complete_df['math_score'] >= 70) 
                                              & (school_data_complete_df['reading_score'] >= 70)]

passing_math_reading.head()                                        
```

```python
total_passing = passing_math_reading['student_name'].count()
total_passing
```

```python
perct_total_passing = total_passing/float(student_count) * 100
print(perct_total_passing)
```

```python
# adding a list of values with keys to create a new DF
district_summary_df = pd.DataFrame(
    [{'Total Schools': school_count,
     'Total Students': student_count,
     'Total Budget': total_budget,
     'Average Math Score': avg_math_score,
     'Average Reading Score': avg_reading_score,
     '% Passing Math': perct_passing_math,
     '% Passing Reading': perct_passing_reading,
     '% Overall Passing': perct_total_passing}])
district_summary_df
```

```python
# Format the "Total Students" to have the comma for a thousands separator.
district_summary_df["Total Students"] = district_summary_df["Total Students"].map("{:,}".format)

district_summary_df["Total Students"]
```

```python
# format total budget to have a comma, $, 2 decimal spaces
district_summary_df['Total Budget']=district_summary_df['Total Budget'].map('${:,.2f}'.format)

district_summary_df
```

```python
# format the columns

district_summary_df['Average Math Score'] = district_summary_df['Average Math Score'].map('{:.1f}'.format)

district_summary_df['Average Reading Score'] = district_summary_df['Average Reading Score'].map('{:.1f}'.format)

district_summary_df['% Passing Math'] = district_summary_df['% Passing Math'].map('{:.0f}'.format)

district_summary_df['% Passing Reading'] = district_summary_df['% Passing Reading'].map('{:.0f}'.format)

district_summary_df['% Overall Passing'] = district_summary_df['% Overall Passing'].map('{:.0f}'.format)
```

```python
# create a grade level DF for math

ninth_graders = school_data_complete_df[(school_data_complete_df['grade'] == '9th')]

tenth_graders = school_data_complete_df[(school_data_complete_df['grade'] == '10th')]

eleventh_graders = school_data_complete_df[(school_data_complete_df['grade'] == '11th')]

twelfth_graders = school_data_complete_df[(school_data_complete_df['grade'] == '12th')]
```

```python
# group each grade level DF by the school name for the avg math score

ninth_grade_math = ninth_graders.groupby(['school_name']).mean()['math_score']

tenth_grade_math = tenth_graders.groupby(['school_name']).mean()['math_score']

eleventh_grade_math = eleventh_graders.groupby(['school_name']).mean()['math_score']

twelfth_grade_math = twelfth_graders.groupby(['school_name']).mean()['math_score']
```

```python
# group each grade level DF by the school name for the avg reading score

ninth_grade_reading = ninth_graders.groupby(['school_name']).mean()['reading_score']

tenth_grade_reading = tenth_graders.groupby(['school_name']).mean()['reading_score']

eleventh_grade_reading = eleventh_graders.groupby(['school_name']).mean()['reading_score']

twelfth_grade_reading = twelfth_graders.groupby(['school_name']).mean()['reading_score']
```

```python
# combine each grade level Series for average math scores by school into a single DF

math_scores_by_grade = pd.DataFrame({
    '9th': ninth_grade_math,
    '10th': tenth_grade_math,
    '11th': eleventh_grade_math,
    '12th': twelfth_grade_math
})

# combine each grade level Series for average reading scores by school into a single DF

reading_scores_by_grade = pd.DataFrame({
    '9th': ninth_grade_reading,
    '10th': tenth_grade_reading,
    '11th': eleventh_grade_reading,
    '12th': twelfth_grade_reading
})
```

```python
math_scores_by_grade['9th'] = math_scores_by_grade['9th'].map('{:.1f}'.format)
math_scores_by_grade['10th'] = math_scores_by_grade['10th'].map('{:.1f}'.format)
math_scores_by_grade['11th'] = math_scores_by_grade['11th'].map('{:.1f}'.format)
math_scores_by_grade['12th'] = math_scores_by_grade['12th'].map('{:.1f}'.format)
#remove the index name
math_scores_by_grade.index.name = None


reading_scores_by_grade['9th'] = reading_scores_by_grade['9th'].map('{:.1f}'.format)
reading_scores_by_grade['10th'] = reading_scores_by_grade['10th'].map('{:.1f}'.format)
reading_scores_by_grade['11th'] = reading_scores_by_grade['11th'].map('{:.1f}'.format)
reading_scores_by_grade['12th'] = reading_scores_by_grade['12th'].map('{:.1f}'.format)
# remove the index name
reading_scores_by_grade.index.name = None
```
